```python
data = pd.DataFrame(eval(file.replace('false', 'False').replace('true', 'True')))
data.to_csv('data.csv')
```

```python
str_lst = '''Economy, jobs, unemployment, wages, optimism, inflation, interest rates, earnings,Stocks market, tariffs, recovery, bull, tax, bear'''.replace(' ', '').lower().split(',')
```

```python
def is_in(string, str_lst = [' data ']):
    
    def my_lable(string, str_lst = str_lst):
        for s in str_lst:
            string = string.replace( s.lower(), ' {%s} '%s.upper())
        return string
    
    string = string.lower()
    
    if any( [(i.lower() in string) for i in str_lst ] ):
        return my_lable(string)
    
    else:
        return np.nan
```

```python
data.columns
```

```python
for v in data['text'].apply(is_in,).dropna().values:
    print(v)
    print('================================================\n')
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_S%26P_500_companies'
stock_df = pd.read_html( requests.get(url).content )[0].T.set_index(0).T
stock_df.head()
```

```python
abbr_lst = stock_df['Symbol'][stock_df['Symbol'].apply(lambda x: len(x) > 1)].tolist()
abbr_lst = [' %s '%i for i in abbr_lst]
abbr_lst
```

```python
for v in data['text'].apply(is_in, str_lst = abbr_lst).dropna().values:
    print(v)
    print('================================================\n')
```
